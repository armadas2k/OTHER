### Сгенерируйте вектор X~(0,1) и постройте гистограмму

```python
# Импортируем нужные библиотеки
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

### Генерировать будет распределение из непрерывных и дискретных величин

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
# создаем массив из 10000 непрерывных значений
array_1 = np.random.uniform(size=10000)
# Создаем серию
array_series_uniform = pd.Series(array_1)

# создаем массив из 10000 непрерывных значений
array_2 = np.random.binomial(1, p=0.7, size=10000)
# Создаем серию
array_series_binomial = pd.Series(array_2)

# Выводим Гистограмму
# Равномерное
sns.histplot(array_series_uniform, ax=ax[0], color='red')
ax[0].set_title('Равномерное распределение', size=15)

#Биномиальное
sns.histplot(array_series_binomial, ax=ax[1])
ax[1].set_title('Биномиальное распределение, p=0.7', size=15)

# Сохраняем в виде png
#plt.savefig('saved_figure.png')
```
